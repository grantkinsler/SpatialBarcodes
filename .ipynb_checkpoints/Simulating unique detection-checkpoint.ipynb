```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('ticks')
plt.style.use("dark_background")
```

```python
n_replicates = 10

# moi_list = np.linspace(1,5,9) # list of target mois to simulate
moi_list = [0.,0.75,1.5,3,6]
moi_list = np.logspace(np.log2(0.05),np.log2(10),num=10,base=2)
thresh_list = [0,1,2] # list of number of unique barcodes in threshold
n_barcodes = 3
n_cells = 10000
```

```python
moi_list
```

```python

frac_passing_out = {}
collision_out = {}

results = {}

for t,target_moi in enumerate(moi_list):
    
    frac_passing_out[target_moi] = {}
    collision_out[target_moi] = {}
    results[target_moi] = {}
    
    for rep in range(n_replicates):

        integration_list = []

        for c in range(n_cells):
            n_integrations = np.random.poisson(target_moi)
            these_integrations = '_'.join([str(x) for x in sorted(np.unique(np.random.choice(range(n_barcodes),n_integrations)))])
            integration_list.append(these_integrations)

        uniques,counts = np.unique(integration_list,return_counts=True)
        uniques = uniques[np.argsort(counts)[::-1]]
        counts = counts[np.argsort(counts)[::-1]]
        n_unique_bcs = np.asarray([u.count('_')+1 if u != '' else 0 for u in uniques ])
        results[target_moi][rep] = {u:count for u,count in zip(uniques,counts)}

#         for u,uni_thresh in enumerate(thresh_list):
            
#             if rep == 0:
#                 frac_passing_out[target_moi][uni_thresh] = []
#                 collision_out[target_moi][uni_thresh] = []
            
# #             frac_passing_out[target_moi][uni_thresh].append(len(counts[np.where(n_unique_bcs > uni_thresh)[0]])/n_cells)
#             frac_passing_out[target_moi][uni_thresh].append(len(counts[np.where(n_unique_bcs > uni_thresh)[0]])/len(counts))
#             collision_out[target_moi][uni_thresh].append(sum(counts[np.where(n_unique_bcs > uni_thresh)[0]]>1)/len(counts[np.where(n_unique_bcs > uni_thresh)[0]]))

        
```

```python
# moi_list = 

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
for u,uni_thresh in enumerate(thresh_list):
    plot_list = []
#     err_list = []s
    for t,target_moi in enumerate(moi_list):
        plot_list.append(np.mean(frac_passing_out[target_moi][uni_thresh]))
    plt.plot(moi_list,plot_list,label=f'nBCs_thresh={uni_thresh+1}',marker='.')
plt.xlabel('Target MOI')
plt.ylabel('Fraction of cells kept after threshold')
plt.legend()
plt.ylim(0,1.01)
        
plt.subplot(1,3,2)
for u,uni_thresh in enumerate(thresh_list):
    plot_list = []
    for t,target_moi in enumerate(moi_list):
        plot_list.append(np.mean(collision_out[target_moi][uni_thresh]))
    plt.plot(moi_list,plot_list,marker='.')
plt.xlabel('Target MOI')
plt.yscale('log')
plt.ylabel('P(collision)')

plt.subplot(1,3,3)
for u,uni_thresh in enumerate(thresh_list):
    plot_list = []
    for t,target_moi in enumerate(moi_list):
        plot_list.append(n_cells*np.mean(collision_out[target_moi][uni_thresh]))
    plt.plot(moi_list,plot_list,marker='.')
plt.xlabel('Target MOI')
plt.yscale('log')
plt.ylabel('Expected number of colliding cells')

plt.suptitle(f'{n_cells} cells, {n_barcodes} barcodes')

plt.tight_layout()
plt.savefig(f'simulated_detection_{n_cells}cells_{n_barcodes}barcodes_DARK.pdf',bbox_inches='tight')
plt.savefig(f'simulated_detection_{n_cells}cells_{n_barcodes}barcodes_DARK.png',dpi=300,bbox_inches='tight')
```

```python
n_replicates = 10

# moi_list = np.linspace(1,5,9) # list of target mois to simulate
moi_list = [0.,0.75,1.5,3,6]
moi_list = np.logspace(np.log2(0.05),np.log2(10),num=10,base=2)
thresh_list = [0,1,2] # list of number of unique barcodes in threshold
n_barcodes = 3
n_cells = 10000
```

```python

frac_passing_out = {}
collision_out = {}

results = {}

for t,target_moi in enumerate(moi_list):
    
    frac_passing_out[target_moi] = {}
    collision_out[target_moi] = {}
    results[target_moi] = {}
    
    for rep in range(n_replicates):

        integration_list = []

        for c in range(n_cells):
            n_integrations = np.random.poisson(target_moi)
            these_integrations = '_'.join([str(x) for x in sorted(np.unique(np.random.choice(range(n_barcodes),n_integrations)))])
            integration_list.append(these_integrations)

        uniques,counts = np.unique(integration_list,return_counts=True)
        uniques = uniques[np.argsort(counts)[::-1]]
        counts = counts[np.argsort(counts)[::-1]]
        n_unique_bcs = np.asarray([u.count('_')+1 if u != '' else 0 for u in uniques ])
        results[target_moi][rep] = {u:count for u,count in zip(uniques,counts)}

#         for u,uni_thresh in enumerate(thresh_list):
            
#             if rep == 0:
#                 frac_passing_out[target_moi][uni_thresh] = []
#                 collision_out[target_moi][uni_thresh] = []
            
# #             frac_passing_out[target_moi][uni_thresh].append(len(counts[np.where(n_unique_bcs > uni_thresh)[0]])/n_cells)
#             frac_passing_out[target_moi][uni_thresh].append(len(counts[np.where(n_unique_bcs > uni_thresh)[0]])/len(counts))
#             collision_out[target_moi][uni_thresh].append(sum(counts[np.where(n_unique_bcs > uni_thresh)[0]]>1)/len(counts[np.where(n_unique_bcs > uni_thresh)[0]]))

        
```

```python
for t,target_moi in enumerate(moi_list):
    this_moi_triples = []
    for rep in range(n_replicates):
        if '0_1_2' in results[target_moi][rep].keys():
            this_moi_triples.append(results[target_moi][rep]['0_1_2']/n_cells)
        else:
            this_moi_triples.append(0)
#         this_moi_triples.append(results[target_moi][rep]['0_1_2']/n_cells)
        
    plt.scatter([target_moi],np.mean(this_moi_triples))
# plt.yscale('log')
plt.xlabel('Average Infections/cell')
plt.ylabel('Fraction of cells with 3 detectable infections')

plt.plot(np.linspace(0.001,10,1000),[x**3/6*np.exp(-x)
                                     for x in np.linspace(0.001,10,1000)],'w--',zorder=0,label='3-infection approximation')

plt.plot(np.linspace(0.001,10,1000),[x**3/6*np.exp(-x)/3+
                                     x**4/(4*3*2)*np.exp(-x)/3*36/81
                                     for x in np.linspace(0.001,10,1000)],'r--',zorder=0,label='3-infection approximation')



# plt.xscale('log')
# plt.yscale('log')
        
```
