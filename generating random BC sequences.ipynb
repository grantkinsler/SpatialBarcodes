```python
import numpy as np
import random
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
```

```python
np.random.seed(9)
```

```python
# WSN repeats

length = 233
n_seqs = 500

seq_list = []

for seq in range(n_seqs):

    sequence = ''.join([np.random.choice(['A','T'])+
            np.random.choice(['G','C'])+
            np.random.choice(['A','T','G','C']) for i in range(length)])
    
    this_record = SeqRecord(
    Seq(sequence),
    id=f"BC_{seq+1:03}",
    name=f"BC_{seq+1:03}",
    description='',
    )
    
    seq_list.append(this_record)
    
    with open(f"WSN_699/all_fastas/BC_{seq+1:03}.fasta", "w") as output_handle:
        SeqIO.write(this_record, output_handle, "fasta")
```

```python
with open("WSN_699/random_sequences.fasta", "w") as output_handle:
    SeqIO.write(seq_list, output_handle, "fasta")
```

```python
## NNNSW

length = 140
n_seqs = 500

seq_list = []

for seq in range(n_seqs):

    sequence = ''.join([
        np.random.choice(['A','T','G','C']) + 
                        np.random.choice(['A','T','G','C']) + 
                        np.random.choice(['A','T','G','C']) + 
        np.random.choice(['A','T'])+
            np.random.choice(['G','C'])
            for i in range(length)])
    
    this_record = SeqRecord(
    Seq(sequence),
    id=f"BC_{seq+1:03}",
    name=f"BC_{seq+1:03}",
    description='',
    )
    
    seq_list.append(this_record)
    
    with open(f"NNNSW_700/all_fastas/BC_{seq+1:03}.fasta", "w") as output_handle:
        SeqIO.write(this_record, output_handle, "fasta")
```

```python
with open("NNNSW_700/random_sequences.fasta", "w") as output_handle:
    SeqIO.write(seq_list, output_handle, "fasta")
```
